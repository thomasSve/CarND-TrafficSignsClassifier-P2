### Pre-process the Data Set (normalization, grayscale, etc.)

Minimally, the image data should be normalized so that the data has mean zero and equal variance. For image data, `(pixel - 128)/ 128` is a quick way to approximately normalize the data and can be used in this project. 

Other pre-processing steps are optional. You can try different techniques to see if it improves performance. 

Use the code cell (or multiple code cells, if necessary) to implement the first step of your project.

````python
### Preprocess the data here. It is required to normalize the data. Other preprocessing steps could include 
### converting to grayscale, etc.
### Feel free to use as many code cells as needed.

import cv2
from sklearn.utils import shuffle

def translate_img(img):
    # Randomly shifts image between -10 and 10 pixels in x and y directions
    rows, cols, _  = img.shape
    shift_x = random.randint(-10, 10)
    shift_y = random.randint(-10, 10)
    M = np.float32([[1,0,shift_x],[0,1,shift_y]])
    img = cv2.warpAffine(img, M, (rows, cols))
    return img[:,:,np.newaxis]

def scale_img(img):
    # Randomly scales image between 1/1.6 and 1.6 without changing shape
    rows, cols, _ = img.shape
    pxl = random.uniform(1/1.6, 1.6)
    pts1 = np.float32([[pxl,pxl],[rows-pxl,pxl],[pxl,cols-pxl],[rows-pxl,cols-pxl]])
    pts2 = np.float32([[0,0],[rows,0],[0,cols],[rows,cols]])
    M = cv2.getPerspectiveTransform(pts1,pts2)
    img = cv2.warpPerspective(img,M,(rows,cols))
    return img[:,:,np.newaxis]

def warp_img(img):
    # Randomly warps image
    rows, cols, _  = img.shape
    
    # Generate three starting point, 1/4 from edges
    x1 = cols / 4
    x2 = x1 * 3
    y1 = rows / 4
    y2 = y1 * 3
    
    # Generate the new position of the three points
    rx = np.random.rand(3) - 0.5 * (cols * 0.06)
    ry = np.random.rand(3) - 0.5 * (rows * 0.06)
    
    pts1 = np.float32([[x1,y1],[x2,y1],[x2,y2]]) # Starting point
    pts2 = np.float32([[x1 + rx[0], y1  + ry[0]], [x2 + rx[1], y1 + ry[1]], [x2 + rx[2], y2 + ry[2]]]) # Decide wrapping end position
    M = cv2.getAffineTransform(pts1,pts2)
    img = cv2.warpAffine(img, M, (rows, cols))
    return img[:,:,np.newaxis]

def random_brightness(img):
    brightness = random.randint(-10, 10) / 128 # Change pixel values between -10 and 10 on normalized image
    img = img[:,:,0] + brightness # Change values of pixel channel
    return img[:,:,np.newaxis]


````

````python
### Testing of the image augmentation methods

methods = [random_brightness, warp_img, scale_img, translate_img]
fig, axes = plt.subplots(len(methods), 2, figsize = (15, 6))
fig.subplots_adjust(hspace = 0.2, wspace = 0.001)

def test_method(method, ax1, ax2):
    i = random.randint(0, len(X_test))
    orig_img = X_test[i]
    new_img = method(orig_img)
    ax1.axis('off')
    ax1.set_title("before {}".format(method.__name__))
    ax1.imshow(cv2.cvtColor(np.squeeze(orig_img), cv2.COLOR_BGR2RGB))
    ax2.axis('off')
    ax2.set_title("after {}".format(method.__name__))
    ax2.imshow(cv2.cvtColor(np.squeeze(new_img), cv2.COLOR_BGR2RGB))

axes = axes.ravel()
i = 0
for m in methods:
    test_method(m, axes[i], axes[i + 1])
    i = i + 2
````

````python
def vis_labels_distribution(X_train, y_train):
    hist, bins = np.histogram(y_train, bins = n_classes)
    width = 0.7 * (bins[1] - bins[0])
    center = (bins[:-1] + bins[1:]) / 2
    plt.bar(center, hist, align='center', width=width)
    plt.show()
````

````python
def image_augmentation(X_train, y_train, min_samples = 750):
    # Creates new samples of images until the min number of samples is reached for each label
    print(n_classes)
    for c in range(n_classes):
        print("Generating samples for class {}".format(c))
        indeces = np.where(y_train == c)
        num_samples = len(indeces[0])
        if num_samples < min_samples:
            for _ in range(min_samples - num_samples):
                img = X_train[random.choice(indeces[0])] # Select random img
                new_sample = translate_img(scale_img(warp_img(random_brightness(img)))) # Generate new sample
                X_train = np.concatenate((X_train, [new_sample]), axis=0)
                y_train = np.concatenate((y_train, [c]), axis=0)
    return X_train, y_train

def preprocess_data(X_train, X_test, y_train, y_test, X_valid):
    # Grayscale the images
    X_train = np.sum(X_train / 3, axis = 3, keepdims = True)
    X_valid = np.sum(X_valid / 3, axis = 3, keepdims = True)
    X_test = np.sum(X_test / 3, axis = 3, keepdims = True)
    
    # Normalize the data to be between (-1, 1)
    X_train = (X_train - 128) / 128
    X_valid = (X_valid - 128) / 128
    X_test = (X_test - 128) / 128
    
    # Use Image augmentation to get a more equal number of image samples for each label
    # This means either scale, translate, rotate, flip, add noise to existing images to generate new ones
    print("Label distribution before augmentation")
    vis_labels_distribution(X_train, y_train)
    
    X_train, y_train = image_augmentation(X_train, y_train)
    
    print("Label distribution after augmentation")
    vis_labels_distribution(X_train, y_train)
    
    # Shuffle the dataset so that the labels are not grouped together
    print("100 first labels before shuffling dataset")
    print(y_train[:100])
    
    X_train, y_train = shuffle(X_train, y_train)
    
    print("100 first labels after shuffling dataset")
    print(y_train[:100])
    
    return X_train, X_test, y_train, y_test, X_valid

X_train, X_test, y_train, y_test, X_valid = preprocess_data(X_train, X_test, y_train, y_test, X_valid)
````

### Model Architecture

````python
### Define your architecture here.
### Feel free to use as many code cells as needed.
import tensorflow as tf
from tensorflow.contrib.layers import flatten

def LeNet(x):
    # Hyperparameters
    mu = 0
    sigma = 0.1
    
    # Layer 1: Convolutional. Input = 32x32x1. Output = 28x28x6
    conv1_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 1, 6), mean = mu, stddev = sigma))
    conv1_b = tf.Variable(tf.zeros(6))
    conv1 = tf.nn.conv2d(x, conv1_W, strides=[1, 1, 1, 1], padding='VALID') + conv1_b
    
    # Activation
    conv1 = tf.nn.relu(conv1)
    
    # Pooling Input = 28x28x6. Output = 14x14x6.
    conv1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    # Layer 2: Convolutional. Output = 14x14x6
    conv2_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 6, 16), mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(16))
    conv2 = tf.nn.conv2d(conv1, conv2_W, strides=[1, 1, 1, 1], padding='VALID') + conv2_b
    
    # Activation
    conv2 = tf.nn.relu(conv2)
    
    # Pooling Input = 10x10x16. Output = 5x5x16.
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    # Flatten Input = 5x5x16. Output = 400.
    fc0 = flatten(conv2)
    
    # Layer 3 Fully Connected. Input = 400. Output = 120.
    fc1_W = tf.Variable(tf.truncated_normal(shape=(400, 120), mean = mu, stddev = sigma))
    fc1_b = tf.Variable(tf.zeros(120))
    fc1 = tf.matmul(fc0, fc1_W) + fc1_b
    
    # Activation
    fc1    = tf.nn.relu(fc1)
    
    # Layer 4: Fully Connected. Input = 120. Output = 84.
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(120, 84), mean = mu, stddev = sigma))
    fc2_b  = tf.Variable(tf.zeros(84))
    fc2    = tf.matmul(fc1, fc2_W) + fc2_b
    
    # Activation
    fc2 = tf.nn.relu(fc2)
    
    # Layer 5: Fully Connected. Input = 84. Output = 10.
    fc3_W  = tf.Variable(tf.truncated_normal(shape=(84, 10), mean = mu, stddev = sigma))
    fc3_b  = tf.Variable(tf.zeros(10))
    logits = tf.matmul(fc2, fc3_W) + fc3_b
    
    return logits
  
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.

x = tf.placeholder(tf.float32, (None, 32, 32, 1))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, 10) # One hot of labels

# Training parameters
learning_rate = 0.001
EPOCHS = 100
BATCH_SIZE = 64

logits = LeNet(x)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = learning_rate)
training_operation = optimizer.minimize(loss_operation)

# Enable to save and load model
saver = tf.train.Saver()

# variables to evaluate model
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

def evaluate_model(X_data, y_data):
    # Evaluates the model on batches and calculates average accuracy    
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
    
    
def train_model(X_train, X_valid, y_train, y_valid):
    print(X_train.shape)
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        num_examples = len(X_train)

        print("Training...")
        for i in range(EPOCHS):
            X_train, y_train = shuffle(X_train, y_train)
            for offset in range(0, num_examples, BATCH_SIZE):
                end = offset + BATCH_SIZE
                batch_x, batch_y = X_train[offset:end], y_train[offset:end]
                sess.run(training_operation, feed_dict={x: batch_x, y: batch_y})

            validation_accuracy = evaluate_model(X_valid, y_valid)
            print("EPOCH {} ...".format(i+1))
            print("Validation Accuracy = {:.3f}".format(validation_accuracy))
            print()

        saver.save(sess, './model')
        print("Model saved")

train_model(X_train, X_valid, y_train, y_valid)
````
